```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/customer_data.csv")
df.head()
```

```python
sns.set(style = "whitegrid")
fig, axes = plt.subplots(nrows =3, ncols = 3, figsize = (14, 16))
sns.boxplot(x="loyalty_status", y = "age", data = df, ax=axes[0,0])
axes[0,0].set_title("Age vs Loyalty_status")
sns.countplot(x="loyalty_status", hue="gender", data = df, ax=axes[0,1])
axes[0,1].set_title("Gender vs Loyalty_status")
sns.boxplot(x="loyalty_status", y="income", data = df, ax=axes[1,0])
axes[1,0].set_title("Income vs Loyalty_status")
sns.countplot(x="loyalty_status", hue="education", data = df, ax=axes[1,1])
axes[1,1].set_title("Education vs Loyalty_status")
sns.countplot(x="loyalty_status", hue="region", data = df, ax=axes[2,0])
axes[2,0].set_title("Region vs Loyalty status")
sns.countplot(x="loyalty_status", hue = "purchase_frequency", data = df, ax= axes[0,2])
axes[0,2].set_title("Purcase frequency vs Loyalty status")
```

```python
sns.set(style="whitegrid")

fig_gender, axes_gender = plt.subplots(nrows=3, ncols=1, figsize=(8, 16))

sns.countplot(x="gender", data=df, hue="purchase_frequency", ax=axes_gender[0], palette="muted")

for p in axes_gender[0].patches:
    axes_gender[0].annotate(format(p.get_height(), '.0f'),
                            (p.get_x() + p.get_width() / 2., p.get_height()),
                            ha = 'center', va = 'center',
                            xytext = (0, 10), textcoords = 'offset points')

axes_gender[0].set_title("Purchase Frequency by Gender")

sns.countplot(x="gender", data=df, hue="product_category", ax=axes_gender[1], palette="muted")

for p in axes_gender[1].patches:
    axes_gender[1].annotate(format(p.get_height(), '.0f'),
                            (p.get_x() + p.get_width() / 2., p.get_height()),
                            ha = 'center', va = 'center',
                            xytext = (0, 10), textcoords = 'offset points')

axes_gender[1].set_title("Product Category by Gender")

sns.countplot(x="gender", data=df, hue="satisfaction_score", ax=axes_gender[2], palette="muted")

for p in axes_gender[2].patches:
    axes_gender[2].annotate(format(p.get_height(), '.0f'),
                            (p.get_x() + p.get_width() / 2., p.get_height()),
                            ha = 'center', va = 'center',
                            xytext = (0, 10), textcoords = 'offset points')

axes_gender[2].set_title("Satisfaction Score by Gender")

plt.tight_layout()

plt.show()

```

```python
df_filtered = df[df['age'] <= 50]

df_filtered['age_group'] = pd.cut(df_filtered['age'], bins=range(20, 55, 5), labels=[f'{i}-{i+4}' for i in range(20, 50, 5)])

sns.set(style="whitegrid")

fig_age, axes_age = plt.subplots(nrows=3, ncols=1, figsize=(12, 16))

sns.countplot(x="age_group", data=df_filtered, hue="purchase_frequency", ax=axes_age[0], palette="muted")
axes_age[0].set_title("Purchase Frequency by Age Group")

sns.countplot(x="age_group", data=df_filtered, hue="product_category", ax=axes_age[1], palette="muted")
axes_age[1].set_title("Product Category by Age Group")

sns.countplot(x="age_group", data=df_filtered, hue="satisfaction_score", ax=axes_age[2], palette="muted")
axes_age[2].set_title("Satisfaction Score by Age Group")

plt.tight_layout()

plt.show()
```

```python
sns.set(style="whitegrid")

fig_region, axes_region = plt.subplots(nrows=3, ncols=1, figsize=(12, 16))

sns.countplot(x="region", data=df, hue="purchase_frequency", ax=axes_region[0], palette="muted")
axes_region[0].set_title("Purchase Frequency by Region")

sns.countplot(x="region", data=df, hue="product_category", ax=axes_region[1], palette="muted")
axes_region[1].set_title("Product Category by Region")

sns.countplot(x="region", data=df, hue="satisfaction_score", ax=axes_region[2], palette="muted")
axes_region[2].set_title("Satisfaction Score by Region")

plt.tight_layout()

plt.show()
```

```python
sns.scatterplot(x="income", y="purchase_amount", data=df)
plt.title("Purchase amount as per Income")
plt.tight_layout()

plt.show()
```
